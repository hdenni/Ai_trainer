```python
import os
import shutil
```

```python
# # 이름이 너무 길어서 줄였음(001_010.json)
filelist = os.listdir(src)
for file in filelist:
    new_name = file[6:9] + '_' + file[19:22] + file[-5:]
    os.rename(src+file, src+new_name)
```
